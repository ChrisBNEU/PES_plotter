# Potential Energy Surface Plotter

The following takes a cantera cti/yaml file and loads it as a gas solution object and a surface Interface object. the species enthalpies and activation energies are plotted using the pes_plot object

uses https://github.com/giacomomarchioro/PyEnergyDiagrams

```python
import cantera as ct
from IPython.display import Image
%matplotlib inline

import os
import sys
sys.path.append(f'{os.getcwd()}/')
import pes_plot
import numpy as np


# get RMG model:
cti_file_path = "chem_annotated.cti"
```

### Create PES plot

```python
ppt = pes_plot.pes_plot(cti_file_path)
```

### Add species to the plot

```python
ppt.add_species("H*(10)", 0)
ppt.add_species("CH3O2*(22)", 0)
ppt.plot()
```

### Add combined species to the plot
allows you to check reactants and their relative enthalpy compared to the single reactant

```python
species = ["CH3O2*(22)","H*(10)"]
ppt.add_combined_species(species,1)
ppt.plot()
```

### add the species to a new position
adds a species to a new position on the PES plot. starts at an arbitrary number, we have chosen 0. 

```python
ppt.add_species("H*(10)", 2)
ppt.add_species("CH3O2*(22)", 2)
ppt.plot()
```

### remove all species, combined species, and reactions from a given position

```python
ppt.purge(2)
ppt.plot()
```

### add a reaction to the plot
specify position and if you want it reversed

```python
ppt.add_reaction("H2*(37) + HCOOH*(19) <=> CH3O2*(22) + H*(10)",2,True)
ppt.plot()
```

### Add the reaction products

```python
species = ["H2*(37)","HCOOH*(19)"]
ppt.add_combined_species(species,3)
ppt.plot()
```

```python
species1 = "H2*(37)"
species2 = "HCOOH*(19)"
ppt.add_species(species1,4)
ppt.add_species(species2,4)
ppt.plot()
```
